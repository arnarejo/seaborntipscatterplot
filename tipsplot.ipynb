```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load data
tips = tips = sns.load_dataset('tips')
print(tips.head())
```

```python
# draw scatter plot in matplotlib.pyplot

plt.scatter(tips.total_bill, tips.tip, color='green', marker='x')
plt.xlabel('total bill')
plt.ylabel('tip')
plt.show()
```

```python
# draw scatter plot in seaborn
sns.regplot(tips.total_bill, tips.tip)
plt.xlabel('total bill')
plt.ylabel('tip')
plt.show()
```
